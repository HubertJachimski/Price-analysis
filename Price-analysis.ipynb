```
#MACBOOK
import requests

from bs4 import BeautifulSoup

macbookPage = requests.get("https://www.ceneo.pl/99105003")

soupMacbook = BeautifulSoup(macbookPage.text, 'html.parser')
macbookStores = soupMacbook.findAll('div', attrs={'class':'product-offer-2020__container'})

macBookStoresNames = []
macBookPrices = []

for i in range (len(macbookStores)):
  macBookPrices.append(macbookStores[i].get('data-price'))
  macBookStoresNames.append(macbookStores[i].get('data-shopurl'))

print(macBookPrices)
print(macBookStoresNames)

file = open('macbooks.csv','w')
for i in range (len(macbookStores)):
    file.write("{} {}\n".format(macBookStoresNames[i], macBookPrices[i]))
file.close()
```

```
import requests
page_g = requests.get("https://best.aliexpress.com/?lan=en")
page_g.text[:200]
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(page_g.text, 'html.parser')

import re
def visible(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title']:
        return False
    elif re.match('<!--.*-->', str(element.encode('utf-8'))):
        return False
    return True

page_text_list = []
for t in filter(visible, soup.findAll(text=True)):
    page_text_list.append(t)

page_text = "".join(page_text_list)
page_text[:100]

page_words = page_text.split()
print(len(page_words), page_words[:10])

import string
page_words2 = [w.strip(string.punctuation).lower() for w in page_words if
len(w.strip(string.punctuation))>0]
print(len(page_words2), page_words2[:100])
```

```
from collections import Counter
page_word_freq = Counter(page_words2).most_common()
print(len(page_word_freq), page_word_freq[:10])
```

```
!pip install wordcloud
%pylab inline
from os import path
import matplotlib.pyplot as plt
from wordcloud import WordCloud
dict_page_word_freq = {}
for i in page_word_freq:
    dict_page_word_freq[i[0]]=i[1]
wordcloud = WordCloud(max_words=200).fit_words(dict_page_word_freq)
plt.imshow(wordcloud)
plt.axis("off")
```
